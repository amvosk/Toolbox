## Example signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000
ntime = 15
scale = np.linspace(0, ntime, ntime*fs, endpoint=False)
signal = np.cos(-scale**2*10)

fig, ax = plt.subplots(1, 1, figsize=(15, 2))
ax.plot(scale, signal)
```

```python
cwt, _ = awt.cwt(signal, freq, decimate=20)
plt.matshow(np.abs(cwt).T, aspect='auto');
```

```python
awt.plot_cwt(signal, freq, decimate=5)
```
